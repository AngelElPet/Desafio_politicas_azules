```python
import numpy as np
import pandas as pd

agua_embalsada = pd.read_csv("csv/evolucion_del_agua_embalsada_en_españa.csv",
                             sep=";")

print(agua_embalsada.head())

agua_embalsada["Año"] = pd.to_datetime(agua_embalsada["Año"], format="%Y")

agua_embalsada.info()
```

```python
agua_embalsada
```

```python
calendario = pd.read_csv("csv/calendar_sunday.csv",
                         sep=",")

calendario["Fecha"] = pd.to_datetime(calendario["Fecha"])

calendario.head()
```

```python
print(agua_embalsada.info())
print(calendario.info())
```

```python
df = pd.merge(agua_embalsada, calendario)
df
```

```python
df.drop(["Año", "Semana Anual", "Semana total", "Día" ], axis='columns', inplace=True)
df
```

```python
df.info()
```

```python
df.to_csv("csv/Agua_embalsada.csv", sep=";")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(20,8))
sns.lineplot(data=df, x="Fecha", y="Agua embalsada")
plt.legend()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.neighbors import KNeighborsRegressor
```

```python
steps = 350
datos_train = df[:-steps]
datos_test  = df[-steps:]

print(f"Filas train : {datos_train.index.min()} a {datos_train.index.max()}")
print(f"Filas test  : {datos_test.index.min()} a {datos_test.index.max()}")

fig, ax = plt.subplots(figsize=(7, 3))
datos_train["Agua embalsada"].plot(ax=ax, label="train")
datos_test["Agua embalsada"].plot(ax=ax, label="test")
ax.legend();
```

```python
X = df["Fecha"].values.reshape(-1,1)
y = df["Agua embalsada"]
```

```python
X_train = X[:1495]
X_test = X[1496:]
y_train = y[:1495]
y_test = y[1496:]
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

LINEAR REGRESION

```python
linreg = LinearRegression()
linreg.fit(X_train, y_train)
```

```python
y_pred = linreg.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
plt.plot(X_test['Fecha'], y_test, label='Real')
plt.plot(X_test['Fecha'], y_pred, label='Predicción')
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.legend()
plt.show()
```

RAMDOM FOREST

```python
rnd_forest = RandomForestRegressor(n_estimators=100)
rnd_forest.fit(X_train, y_train)
y_pred = rnd_forest.predict(X_test)
```

```python
plt.plot(df["Fecha"], df["Agua embalsada"], label='Datos reales')
plt.plot(y_pred, label='Predicciones')
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.legend()
plt.show()
```

KNN

```python
knn = KNeighborsRegressor(n_neighbors=3)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
```

```python
plt.plot(df["Fecha"], df["Agua embalsada"], label='Datos reales')
plt.plot(X_test, y_pred, label='Predicciones')
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.legend()
plt.show()
```

```python
def prediction(model, x_test):
    preds = x_test[0]
    preds_out = []
    
    for i in range(len(X_test)):
        pred = model.predict([preds])[0]
        preds_out.append(pred)
        
        preds2 = np.array([preds[1:]]) # quitamos t-12 ya no se usa
        
        preds = np.c_[preds2, np.array([pred])][0]
        
    return preds_out
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

pred = lin_reg.predict(X_test)
print("MSE:", mean_squared_error(pred, y_test))

pred2 = prediction(lin_reg, X_test) # predigo uno a uno
print("MSE:", mean_squared_error(pred2, y_test))
```
