```python
df["¿Qué son las políticas azules?"] = df["¿Qué son las políticas azules?"].replace \
    ({"Las políticas relacionadas con el agua": 10, 
    "Las políticas relacionadas con los colores": 1})
df["¿Qué Ministerio gestiona las políticas relacionadas con el agua?"] = df["¿Qué Ministerio gestiona las políticas relacionadas con el agua?"].replace \
    ({"Transición Ecológica": 10, 
    "Interior": 1 ,
    "Exteriores": 5,
    "Consumo": 2})
df["¿Quién tiene la titularidad de la gestión del agua?"] = df["¿Quién tiene la titularidad de la gestión del agua?"].replace \
    ({"Titularidad pública": 10,
    "Titularidad privada": 1})
df["En las últimas dos décadas ¿cuántos litros de agua crees que consume de media una persona en su casa?"] = df["En las últimas dos décadas ¿cuántos litros de agua crees que consume de media una persona en su casa?"].replace \
    ({"33,8": 5,
    "53,4": 2,
    "39,3": 10,
    "68,4": 1})
df["En mayo de 2023 ¿cuál es la comunidad autónoma que tiene los embalses más vacíos?"] = df["En mayo de 2023 ¿cuál es la comunidad autónoma que tiene los embalses más vacíos?"].replace \
    ({"Extremadura": 2,
    "Cataluña": 5,
    "Asturias": 1,
    "Andalucía": 10})
df["¿Cuál es la cuenca que más embalses tiene en España?"] = df["¿Cuál es la cuenca que más embalses tiene en España?"].replace \
    ({"Cuenca del Guadalquivir": 2,
    "Cuenca del Tajo": 5,
    "Cuenca del Segura": 1,
    "Cuenca del Ebro": 10})
df["¿Cuál es el porcentaje de reutilización de aguas residuales tratadas en España?"] = df["¿Cuál es el porcentaje de reutilización de aguas residuales tratadas en España?"].replace \
    ({"25%": 2,
    "5%": 5,
    "50%": 1,
    "9,96%": 10})
df["¿Cómo crees que evolucionarán el estado de los embalses en el futuro?"] = df["¿Cómo crees que evolucionarán el estado de los embalses en el futuro?"].replace \
    ({"Empeorará": 10,
    "Mejorará": 5})
```

```python
df_1 = df.drop(['Marca temporal', 'Elige la comunidad autónoma en la que vives:'], axis=1)
df_1
```

```python
import pandas as pd
import numpy as np

def replicate_data(dataframe, num_replicas):
    replicated_data = []
    for _, row in dataframe.iterrows():
        for _ in range(num_replicas):
            replicated_data.append(row)
    return pd.DataFrame(replicated_data, columns=dataframe.columns)

# Ejemplo de uso
original_data = df_1
replicated_data = replicate_data(original_data, num_replicas=3)
```

```python
# Concatenating DataFrames
combined_data = pd.concat([original_data, replicated_data], ignore_index=True)

print("Datos combinados:")
print(combined_data.shape)
```

```python
data_score = combined_data.assign(score=combined_data.sum(axis=1))

print("DataFrame con columna 'score':")
print(data_score.score)
```
