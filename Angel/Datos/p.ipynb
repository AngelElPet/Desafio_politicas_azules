# Hogares

Volumen en miles de metros c煤bicos al d铆a

```python
print(hog.info())
```

```python
print(hog.describe())
```

```python
hog.head()
```

```python
hog = hog.iloc[1:,:]
```

```python
Media = []
a=0
for y in range(len( hog['Provincia'])):
    for x in range(len(hog.columns)-1):
        a =a +hog.iloc[y,x+1]
    Media.append(a/len(hog['Provincia']))
    a=0

hog['Media'] = Media
```

```python
import matplotlib.pyplot as plt
for t in hog.columns[1:]:
    hog = hog.sort_values(t,ascending=True)
    hog.head()
    x = hog[t]

    plt.figure(figsize=(7,7))
    plt.hlines(y = hog['Provincia'], xmin = min(x)-10000, xmax= max(x), color ='skyblue')
    plt.title('A帽o ' + t)
    plt.plot(x,hog['Provincia'],'o');
```
