```python
import pandas as pd
import matplotlib.pyplot as plt

capacidad_embalses = pd.read_csv("C:/Users/Didac/Desktop/AGUA/csv/asi_ha_evolucionado_el_volumen_y_la_capacidad_de_los_embalses_en_españa.csv", sep=";")
print(capacidad_embalses.info())
print(capacidad_embalses.head())



# Elimino los nulos.
capacidad_embalses = capacidad_embalses.dropna(subset=['Año', 'Agua embalsada', 'Capacidad de los embalses'])

# Paso año a int
capacidad_embalses['Año'] = capacidad_embalses['Año'].astype(int)

# Restablece el índice del DataFrame
capacidad_embalses.reset_index(drop=True, inplace=True)


plt.figure(figsize=(10, 6))

plt.plot(capacidad_embalses['Año'], capacidad_embalses['Agua embalsada'], label='Agua embalsada')
plt.plot(capacidad_embalses['Año'], capacidad_embalses['Capacidad de los embalses'], label='Capacidad de los embalses')

plt.title('Agua Embalsada y capacidad')
plt.xlabel('Año')
plt.ylabel('Volumen')
plt.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


consumo_medio_CCAA = pd.read_csv("C:/Users/Didac/Desktop/AGUA/csv/consumo_medio_de_agua_de_los_hogares_por_comunidad_autonoma.csv",
                                 sep=";")

consumo_medio_CCAA = consumo_medio_CCAA.dropna()



consumo_medio_CCAA = consumo_medio_CCAA.sort_values(by=['2018', '2020'], ascending=False)

plt.figure(figsize=(10,6))
bars_2018 = plt.barh(consumo_medio_CCAA['Parámetro'], consumo_medio_CCAA['2018'], color='blue', label='2018')
bars_2020 = plt.barh(consumo_medio_CCAA['Parámetro'], consumo_medio_CCAA['2020'], left=consumo_medio_CCAA['2018'], color='slateblue', label='2020')

plt.title('Consumo medio de agua por hogar por comunidad autónoma')
plt.xlabel('Consumo medio de agua')
plt.ylabel('Comunidad Autónoma')
plt.legend()

# Valores de las columnas
for bar in bars_2018:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f'{bar.get_width():.0f}', 
             va='center', ha='right', color='white', fontsize=8)

for bar in bars_2020:
    plt.text(bar.get_width() + bar.get_x(), bar.get_y() + bar.get_height()/2, f'{bar.get_width():.0f}', 
             va='center', ha='right', color='white', fontsize=8)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargamos los datos
embalses_cuenca = pd.read_csv("C:/Users/Didac/Desktop/AGUA/csv/Embalses_cuenca.csv", sep=",")
embalses_cuenca['2022'] = embalses_cuenca['2022'].str.replace(',', '.').astype(float)
embalses_cuenca['2012'] = embalses_cuenca['2012'].str.replace(',', '.').astype(float)

# Ordenamos los datos por el volumen de embalses en 2022
embalses_cuenca = embalses_cuenca.sort_values(by='2022', ascending=False)

plt.figure(figsize=(10,6))

# Barras de 2022
bars_2022 = plt.barh(embalses_cuenca['Cuencas'], embalses_cuenca['2022'], color='blue', label='2022')

# Barras de 2012, desplazadas a la derecha
bars_2012 = plt.barh(embalses_cuenca['Cuencas'], embalses_cuenca['2012'], left=embalses_cuenca['2022'], color='slateblue', label='2012')

plt.title('Volumen de embalses por cuenca')
plt.xlabel('Volumen de embalses')
plt.ylabel('Cuenca')
plt.legend()

# Valores de las columnas
for bar in bars_2022:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f'{bar.get_width():.1f}', 
             va='center', ha='right', color='white', fontsize=8)

for bar_2022, bar_2012 in zip(bars_2022, bars_2012):
    plt.text(bar_2022.get_width() + bar_2012.get_width(), bar_2022.get_y() + bar_2022.get_height()/2, f'{bar_2012.get_width():.1f}', 
             va='center', ha='right', color='white', fontsize=8)

plt.show()




```

```python
import pandas as pd
import matplotlib.pyplot as plt

suministro_saneamiento_CCAA = pd.read_csv("C:/Users/Didac/Desktop/AGUA/csv/suministro_y_saneamiento_agua_CCAA_2000-2020", sep=",")

# Total nacional me lo cargo
suministro_saneamiento_CCAA = suministro_saneamiento_CCAA[suministro_saneamiento_CCAA['Provincia'] != 'Total Nacional']

df = suministro_saneamiento_CCAA.set_index('Provincia').T

# Crear una lista de colores para las líneas
colors = plt.cm.tab20(range(len(df.columns)))


plt.figure(figsize=(10, 6))
lines = df.plot(kind='line', color=colors)

plt.title('Suministro y saneamiento de agua por comunidad autónoma')
plt.xlabel('Año')
plt.ylabel('Cantidad')

plt.legend(title='Comunidad Autónoma', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```
